```python
# choose a key to update for all the configs
import json
import os 

def update_config(path, key, value, action):
    assert action in ['add', 'update', 'remove']
    configs = [file for file in os.listdir(path) if file.endswith('.json')]
    for file in configs:

        config = json.load(open(file, 'r'))
        
        if "task_name"  not in config:
            continue
        
        with open(file, 'w') as f:
            if action == 'update':
                if key in config:
                    config[key] = value
                    json.dump(config, f, indent=4)
            elif action == 'add':
                config[key] = value
                json.dump(config, f, indent=4)
            else:
                del config[key]
                json.dump(config, f, indent=4)
        
```

```python
# add a key to all the configs
ACTIONS = ['add', 'update', 'remove']

key = 'save_steps'
value = 150
action = 'add'
path = "."

update_config(path, key, value, action)
```
